```python
var('x')
f(x) = sin(x)*x + 4
P = line([(2,0),(2,f(2))], color="black")
P += line([(8,0),(8,f(8))], color="black")
P += polygon( [(2,0),(2,f(2))] + [(x, f(x)) for x in [2,2.1,..,8]] + [(8,0),(2,0)], 
             rgbcolor=(0.5,0.8,0.7), aspect_ratio='auto')
P += text("$\\int_{2}^8 (sin(x) x+4) dx$", (4.5, 8), fontsize=14, color="black")
P += plot(f, (x, 1.5, 8.5), thickness=2, figsize=5)
show(P)
```
